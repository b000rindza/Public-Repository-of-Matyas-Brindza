# Oszcillátorok numerikus vizsgálata

## Először ismételjük át az alapokat:

Oszcillátoroknak olyan dinamikai rendszereket nevezünk, melyek mozgása az idő periódikus vagy kvázi-periódikus függvényeként írható le. Az egydimenziós oszcillátor mozgásegyenlete

$$ m \frac{\dd^2x}{\dd t^2} = - \frac{\dd V(x)}{\dd x} $$

alakban írható. Egy szabad oszcillátor teljes energiája mindaddig megmarad, amíg a rendszerben nem hat súrlódási erő (csillapított oszcillátor), illetve nem hatunk rá külső erővel (gerjesztett oszcillátor). Az oszcillátor teljes energiája a kinetikus és a potenciális energia összege, azaz

$$ E_{\textrm{tot}} = \frac{m v^2}{2} + V(x)$$

### Harmonikus oszcillátor

Amennyiben a $V(x)$ potenciál négyzetes, abban az esetben harmonikus oszcillátorról beszélünk:

$$\begin{eqnarray}
    V(x) & = & \frac{1}{2}kx^2 \\
    m \frac{\dd^2x}{\dd t^2} & = & - k x.
\end{eqnarray}$$

A harmonikus oszcillátor mozgásegyenlete analitikusan is megoldható. A megoldás

$$ x(t) = A \cos( \omega t + \varphi ) $$

alakú, ahol $A$ az amplitúdó - vagyis a legnagyobb kitérés -, $\omega$ a körfrekvencia és $\varphi$ a kezdőfázis $t = 0$-ban. A harmonikus oszcillátor körfrekvenciáját a potenciálban szereplő $k$ állandó és az $m$ tömeg jellegű mennyiség határozza meg:

$$ \omega = \sqrt{\frac{k}{m}} = \frac{2\pi}{T}, $$

ahol $T$ a pediódusidő, melyből $f = 1/T$ módon kaphatjuk meg a frekvenciát.

## Anharmonikus oszcillátor

Az anharmonikus abban különböznek a harmonikus oszcillátortól, hogy a mozgásukat meghatározó potenciál nem négyzetes alakú. Néhány lehetséges anharmonikus potenciál:

* Negyedrendű potenciál

$$ V(x) = \frac{1}{4}k_4 x^4 $$

* Lennard-Jones potenciál

$$ V(x) = V_0 \left[ \left( \frac{x_0}{x} \right)^{12} - 2 \left( \frac{x_0}{x} \right)^{6} \right] $$

## Csillapított és gerjesztett oszcillátor

A csillapított és/vagy gerjesztett oszcillátorok mozgásegyenletében megjelenik egy sebességtől függő és egy további, többnyire időfüggő külső erőtag is:

$$ m \frac{\dd^2x}{\dd t^2} = - \frac{\dd V(x)}{\dd x} - c \frac{\dd x}{\dd t} + F(t) $$

Csillapított és szinuszosan gerjesztett oszcillátorok viselkedése analitikusan is egyszerűen vizsgálható. Az alábbi feladatokban ezért ezektől eltérő rendszereket vizsgálunk numerikus módszerek segítségével.

## 2. feladatrész

Integráljuk a fentiek közül kettő rendszer, mégpedig a harmonikus és a Lennard-Jones potenciállal meghatározott oszcillátorok mozgásegyenletét a python scipy.integrate.RK45 csomagjának használatával! Ez egy adaptív Runge-Kutta eljárás, azaz a léptetés nagyságát hibabecslés alapján állítja be.

A scipy.integrate.RK45 használata sokban hasonlít a fiznum1-ből tanult odeint használatához. A lényegesebb argumentumok ezek:

rk45 = RK45( 'fun', 't0', 'y0', 't_bound', 'max_step=inf', 'rtol=0.001', 'atol=1e-06' )

Itt 'fun' a differenciálegyenletet leíró függvény neve, 't0' a kezdő időpont, 'y0' a kezdő állapot.

Lényeges eltérés viszont az odeint függvénytől, hogy az RK45 függvényhíváskor még csak egy objektumot kapunk eredményül.
Ennek .step metódusát tudjuk az integrálási lépésekre használni, a .t, .y attribútumaiként pedig megkapjuk a változók értékeit. Tehát ha az RK45 eredményét a fentiek szerint eltároljuk, akkor a léptetés rk45.step() utasítással végezhető, az eredményeket pedig rk45.t, rk45.y értékeiként kapjuk meg.

Ábrázoljuk a harmonikus és a Lennard-Jones potenciállal rendelkező rendszerek változóinak időfejlődését, a teljes energiát és a lépéshosszt az idő függvényében oszcilláló mozgás esetén néhány periódus hosszan!

Megfelelő pontossághoz az rtol értéket 1e-5-re csökkentsük!
Továbbá a Lennard-Jones potenciál esetén indítsuk a rendszert az egyensúlyi helyzetből pozitív irányban $x_0/3$ értékkel kitérítve, kezdősebesség nélkül. Mutassuk ki, hogy a lépéshossz jelentősen változik időben! Van-e lényeges lépéshossz változás a harmonikus oszcillátornál is?

```python
from scipy.integrate import RK45
```

```python
kezdvx = [1.5,0] # a "kezd" meghatározza a kezdeti helyzetet, és sebességet
m = 2 # a test tömege
k = 72
t_bound = np.inf

def energia(potencial,m,x,v):
    pot,f = potencial(x) # potenciális energia
    kin = 0.5 * m * v**2 # kintetikus energia
    return pot + kin # össz energia


# ezek a függvények írják le a difegyenleteket
def kf_har(t,u):
    # kezdeti helyzet
    x=u[0]
    # gyorsulás
    # F = -k*x = m*a
    a=-k*u[1]/m
    return [a,x] 

def kf_negy(t,u):
    # kezdeti helyzet
    x=u[0]
    # gyorsulás
    # F = -k*x**3 = m*a
    a=-k*(u[1])**3/m
    return [a,x] 



def kf_lj(t,u):
    # kezdeti helyzet
    x=u[0]
    # gyorsulás
    # a = F/m
    a=12*V0*(x0**12/u[1]**13-x0**6/u[1]**7)/m
    return [a,x]
    

def integral(kf,kezdvx,step_num,rtolbe = 1e-5, atole = False, step_max = np.inf):
    # nem adaptívnál kell majd piszkálni az atol paramétert
    # egyéb esetben maradhat a default 1e-6
    if atole==False:
        teszt = RK45(kf,0,kezdvx,t_bound,rtol = rtolbe, max_step = step_max) # integrál létrehozása
    if atole==True:
        teszt = RK45(kf,0,kezdvx,t_bound,rtol = rtolbe, atol = atolbe, max_step = step_max) # integrál létrehozása
    vx = [] # ide jönnek a dupla ODE integrált értékei
    ido = [] # időpillanatok
    step_size = [] # lépéshosszok
    for i in range(step_num):
        vx.append(teszt.y) # ide kerülnek az integrálás értékei
        ido.append(teszt.t) # ide kerülnek az időpillanatok
        step_size.append(teszt.step_size) # ide kerülnek a lépéshosszok
        teszt.step() # léptetés
    return vx, ido, step_size
```

```python
# integrálás és ábrázolás egyben
# ez ezért kényelmes, mert csak 2 argumentuma lesz a függvénynek
# ( további három "if" beiktatásával lehetne csupán egy paramétere is )
def abra_int(pot,kf):
    # adatok kigenerálása
    u, ido, step_size = integral(kf,kezdvx,step_num) # elvégzem az integrálást
    u = np.array(u) # listát nem lehet tuple-lel indexelni, csak np tömböt
    en_ossz = np.array(energia(pot,m,u[:,1],u[:,0])) # össz energia
    

    
    # ábrázolás
    fig, axs = plt.subplots(2, 2,figsize=(12,12))
    if pot==har:
        s = "Harmonikus oszcillátor"
    if pot==negyed:
        s = "Negyedrendű potenciál"
    if pot==lj:
        s = "Lennard-Jones potenciál"
    plt.suptitle(s,fontsize=25)
    fig.tight_layout(pad=6.0)

    axs[0, 0].plot(ido,np.array(u)[:,1],label="x[t]")
    axs[0, 0].set_title("Helyzet")
    axs.flat[0].set(xlabel="t[s]",ylabel="x[m]")

    axs[0, 1].plot(ido,np.array(u)[:,0],label="v[t]")
    axs[0, 1].set_title("Sebesség")
    axs.flat[1].set(xlabel="t[s]",ylabel="v[m/s]")

    axs[1, 0].plot(ido,en_ossz,label="E(t)")
    axs[1, 0].set_title("Energia")
    axs.flat[2].set(xlabel="t[s]",ylabel="E[J]")
    axs.flat[2].set(ylim=(0,en_ossz[0]*1.2))
    axs.flat[2].ticklabel_format(useOffset=False, style='plain', axis='y')
 
    axs[1, 1].plot(ido,step_size)
    axs[1, 1].set_title("Lépéshosszok")
    axs.flat[3].set(xlabel="t[s]",ylabel="$\\tau$[s]")
```

```python
step_num = 100
kezdvx = [0,1]
abra_int(har,kf_har)
```

```python
kezdvx = [0,1+x0/3]
abra_int(lj,kf_lj)
```

Míg a harmonikus oszcillátor esetén 0.06 nagyságú lépések vannak, az LJ-potenciál esetén másfél nagyságrenddel nagyobb lépéshosszokat látunk - sőt a lépéshosszok eloszlása is különbözik :
- harmonikus oszcillátor : $0.06\pm 0.005$
- LJ-potenciál : $0.125\pm 0.1$

## 3. feladatrész

Vizsgáljuk a Lennard-Jones-potenciál esetén, hogy az adaptív, a nemadaptív Runge-Kutta és az Euler módszer hibája hogyan viszonyul egymáshoz, ha azonos lépésszámot próbálunk elérni! Mivel a rendszer energiája elméletileg konstans, az energiának a kezdeti értéktől való eltérését használhatjuk a hiba jelzésére, azt ábrázoljuk az idő függvényében néhány perióduson keresztül!

Az egyes léptetésekben az adaptív eljárással azonos, de ugyanakkor nem adaptív megoldást úgy tudunk létrehozni, hogy az RK45 eljárás használatakor a hibahatárokat (az atol és rtol nevű opcionális paramétereket) 0.1-re állítjuk, ekkor az a hibabecslése alapján viszonylag nagy lépéseket használna. Ugyanakkor a lépések nagyságát a max_step nevű opcionális paraméterrel korlátozva gyakorlatilag kikapcsoljuk a saját lépésköz szabályzását. Használjuk itt lépésköznek azt az értéket, amivel ugyanannyi lépés jön létre az eddig használt időintervallumon, mint amennyi az adaptív esetben volt!

Az Euler módszerre írjunk egyszerű for ciklust, vagy függvényt, amiben az előbbi lépésközt használjuk! Ábrázoljuk a három módszerrel az x koordinátára kapott eredményeket egy ábrában, és egyben pedig az energia megváltozását a kezdetihez képest. Amennyiben valamelyik eljárás nem adna oszcilláló megoldást, annál addig csökkentsük a lépésközt 10-ed, 100-ad részére, amíg oszcilláló lesz! Ha pedig az energiaváltozás ábráján nem látszik egyszerre a 3 esetben az energiaváltozás, akkor használjunk logaritmikus skálát.

```python
step_num=150
kezdvx = [0,1+x0/3]
```

```python
# adaptív Runge-Kutta

# def integral(kf,kezdvx,step_num,rtolbe = 1e-5, atole = False, step_max = np.inf):
vx, ido_ad, step_size_ad = integral(kf_lj,kezdvx,step_num,rtolbe=1e-5,atole = False) # integrálás
vx_ad = np.array(vx)
x_ad = vx_ad[:,1] # kitérés
energia_ad = np.array(energia(lj,m,vx_ad[:,1],vx_ad[:,0])) # energia
```

```python
# nem adaptív Runge-Kutta

atolbe = 0.1 # abszolút tolerancia
step_max = ido_ad[-1]/step_num #!!!
h = step_max
k = 1 #ennyiszer kisebb lépéshossz
step_nmax = step_max/k
step_n = step_num*k
# def integral(kf,kezdvx,step_num,rtolbe = 1e-5, atole = False, step_max = np.inf):
#integrálás
vx, ido_nad, step_size_nad = integral(kf_lj,kezdvx,step_num,rtolbe=0.1,atole = True,step_max=h)
vx_nad = np.array(vx)
x_nad = vx_nad[:,1] # kitérés
energia_nad = np.array(energia(lj,m,vx_nad[:,1],vx_nad[:,0])) # energia
```

```python
# Az Euler-módszert megvalósító függvény
def euler(fun,t0,x0,v0,h,step_num):
    v = [v0]
    x = [x0]
    t = [t0]
    for i in range(1,step_num+1):
        V , F = fun(x[-1])
        a = F/m
        #print(a)
        t.append(t0+i*h)
        v.append(v[-1]+h*a)
        x.append(x[-1]+h*v[-1])
    return np.array(t), np.array(x), np.array(v)
```

```python
# Euler-módszer adatok kigenerálása
ido_eu, x_eu, v_eu = euler(lj,0,kezdvx[1],kezdvx[0],step_max,step_num)
energia_eu = np.array(energia(lj,m,x_eu,v_eu))
```

## 4.a) feladatrész

A következő feladatrészben Fourier spektrumot kell számolnunk, amit közvetlenül csak időben egyenletes lépésközű adatokon kaphatunk meg. Az adaptív lépésközű integrálás bár hasznos, azonban a koordináta értékek nem egyenletes lépésközzel adódtak ki. Ezért interpoláljuk a 3. feladatból kapott megoldásokat időben egyenletes lépésközzel, lineárisan és köbös spline-ok segítségével!

Az interpolációhoz a scipy.interpolate csomag interp1d() függvényének használata javasolt.
Ha van egy t listában, vagy arrayben tárolt időpont sorozatunk és egy y listában/arrayben a hozzájuk tartozó függvényértékek, akkor ezekhez az alábbi módon tudunk interpolációt végezni:

ill=interp1d(t,y)

Az eredmény még nem adatsor lesz, hanem egy függvény, amit ill(t1) módon meghívva kapjuk az illesztett függvény értékeit. Alapból az illesztés lineáris, tehát a beadott adatsorok értékeiből alkotott szakaszonként egyenes függvény t1-nél felvett értékét kapjuk meg. A 'cubic' opciót az interp1d-ben megadva az illesztés köbös lesz.
Az eredményül kapott ill függvénynek, mint legtöbb python függvénynek egyszerre több értéket is adhatunk egy listában, vagy arrayben.

Hasonlítsuk össze a lineáris és köbös illesztés eredményét!

## <font color='green5'>5. Szorgalmi feladatrész</font>

Programozzunk be kaotikus oszcillátort!$^1$ Egy szinuszosan gerjesztett oszcillátor esetében a kaotikus viselkedés feltétele, hogy a visszatérítő erő ne lineáris függvénye legyen a kitérésnek. Ábrázoljuk a kitérés és a sebesség időfüggését, valamint a teljesítményspektrumot! Futassuk az integrálást legalább 1 millió lépésig, és ábrázoljuk a rendszer Poincaré-metszetét, azaz a sebességet a kitérés függvényében az $\omega t = n \cdot 2 \pi, n = 1, 2, 3, ... $ feltétel mellett!

$^1$ http://sprott.physics.wisc.edu/pubs/paper265.pdf

```python
# a megoldás helye
```

```python
# kaotikus oszcillátort difegyenletét megvalósító függvény
def kf_chaos(t,u):
    x=u[0] # az első komponens x deriváltja
    v=-np.sign(u[1])*k*np.abs(u[1])**p/m+np.sin(omega*t) # a második komponens v deriváltja
    return  [v,x] 
```

```python
# ez a függvény megvalósítja a fenti integrálást és a Poincaré metszethez szolgáltat adatokat
def integral_poincare(kf,kezdp,step_num,rtolbe = 0.001, mx_step = np.inf):
    teszt = RK45(kf,0,kezdp,t_bound = np.inf,rtol = rtolbe, max_step = mx_step)
    n = 1
    u = []
    ido = []
    step_size = []
    poincare_u = []
    for i in range(step_num):
        u.append(teszt.y)
        ido.append(teszt.t)
        step_size.append(teszt.step_size)
        #if (round(omega*teszt.t,2) % round(2*np.pi,2) == 0): # explicit közelítés
        if (omega*teszt.t >= n*2*np.pi): # dinamikus közelítés  
            poincare_u.append(teszt.y)
            n += 1
        teszt.step()
    return u, ido, step_size, poincare_u

# Ezt a függvényt végül nem használtam, mert túl sok false positive és false negative pont lett
# Az volt a tapasztalatom, hogy a "dinamikus közélítés" feltétele jobb volt,
# mint az explicit közelítésé

# Az explicit közelítéssel végzett összehasonlításban egy paraméternek muszáj (t[1]-t[0])-tól függnie,
# különben más t tömbökre más lesz a hibatűrés.
# Ezt 2 módon lehetne megoldani:
# 1) sokat próbálgatjuk és manuálisan beállítjuk
# 2) sokat próbálgatjuk, analitikusan számolunk és képlet formájában megadjuk a (t[1]-t[0]) függést
# A modulus is egy hibaforrás

# A dinamikus közelítést olyan elgondolásból próbáltam meg, hogy t lépésközei kicsik,
# legalábbis sokkal kisebbek, mint 2*pi, ezért ha épp nagyobb t értéke mint 2*pi-nek egy többszöröse,
# akkor jó közelítéssel egyenlőek is. Az n változó léptetett.
```

```python
# paraméterek
kezdp = [0,0]
omega = 1.88
step_num = 1000000
m = 1
k = 1
p = 3
```

```python
# adatok kigenerálása
vx_chaosr, ido_chaosr, step_chaosr, poincare_chaos = integral_poincare(kf_chaos,kezdp,step_num,rtolbe = 1e-05)
vx_chaosr = np.array(vx_chaosr)
poincare_chaos = np.array(poincare_chaos)
```
