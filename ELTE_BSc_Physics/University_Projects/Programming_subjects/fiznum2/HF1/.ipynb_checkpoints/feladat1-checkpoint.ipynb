# Passzív áramkörök numerikus vizsgálata

Lineáris, időben állandó fizikai rendszerek vizsgálatakor gyakran foglalkozunk ideális elektronikus áramkörökkel, hiszen az elméleti érdekességeken túl ezek a gyakorlatban is igen fontosak. Az alábbi feladatok passzív elektronikai alkatrészekből (ellenállás, kondenzátor és induktivitás) megépíthető áramkörök viselkedését vizsgálják, de az alkalmazott módszerek a jelfeldolgozás területén általánosan használhatók. A feladatok megoldása során az `ahkab` áramkörszimulációs csomagot használjuk, mely nagy időfelbontással számítja ki az áramkör egyes csomópontjaiban megjelenő jelalakokat (időfüggő feszültségeket). Az ilyen szimuláció közelebb áll a valódi analóg áramkörök működéséhez, mint más, ún. diszkrét idejű rendszerek. Ezutóbbiak a digitális jelfeldolgozás modelljének tekinthetők.

A passzív elektronikai altrészek viselkedése a következő egyszerű fizikai törvényekkel írhatók le, ahol $u = u(t)$ az időfüggő feszültség, $i = i(t)$ az áram, $R$ az elleneállás, $C$ a kapacitás és $L$ az induktivitás:

* Ellenállás: $$ u = R i $$

* Kondenzátor: $$ \frac{du}{dt} = \frac{1}{C} i $$

* Induktivitás: $$ u = L \frac{di}{dt} $$

Konkrét áramkörök esetében az egyes komponensekre felírt differenciálegyenleteket a [Kirchhoff-törvények](https://hu.wikipedia.org/wiki/Kirchhoff-törvények) kapcsolják rendszerbe. Az egyenletek megoldásakor valamelyik bemeneti jelalakot (feszültség- vagy áramérték időbeli változását) előírjuk, és a rendszer erre adott válaszát, azaz egy kimeneti jelalak időfejlődését keressük.

Érdekes elméleti kérdés, hogy egy rendszer milyen választ ad egy nagyon rövid ideig tartó vizsgálóimpulzus esetében. Matematikailag az ilyen vizsgálóimpulzusokat Dirac-delta függvénnyel írjuk le, és ismert, hogy egy lineáris, időinvariáns rendszer válaszfüggvénye tetszőleges gerjesztés esetében meghatározható az **impulzusválasz** (más néven Green-függvény) és a gerjesztés konvolúciójaként:

$$
u_\mathrm{out}(t) = \int_{-\infty}^{t} u_\mathrm{in}(\tau) g(t-\tau) d\tau,
$$

ahol az, hogy az integrálás $t$-ig tart a rendszer kauzalitását fejezi ki.

Lineáris időinvariáns rendszereket érdemes Fourier-kifejtésben vizsgálni. Ilyenkor a rendszer szinuszos bemenő gerjesztésekre adott válaszát keressük, ami a tranziens viselkedés lecsengése után a linearitás miatt azonos frekvenciájú, szintén szinuszos jelalak lesz. Egy rendszer **átviteli függvénye** megadja, hogy egy $f$ frekvenciájú szinuszos gerjesztés esetében a rendszer mekkora csillapítást, illetve fázistolást okoz. Ismert, hogy Fourier-térben a konvolúció művelete a gerjesztés és az impulzusválasz Fourier-transzformáltjának szorzataként írható fel, és ebből az is következik, hogy az átviteli függvény az impulzusválasz Fourier-transzformáltja.

A feladatok során passzív áramköri elemekből összállított analóg szűrőket vizsgálunk, melyek az ún. Butterworth-topológiát követik. Az interneten elérhetők olyan eszközök, melyekkel különböző átviteli karakterisztikájú passzív vagy aktív szűrők tervezhetők, ld. pl.:

## Előre definiált függvények

Az alábbi cellák olyan függvényeket töltenek be és definiálnak, melyek segítségével az `ahkab` programcsomag részletes ismerete nélkül is kiszámíthajuk az áramkörök bemenő jelre adott válaszfüggvényét.

Az első függvénydefiníció a `run_transient` függvényé, mellyel majd a szimulációt tudjuk elvégezni. Utána mindhárom példa áramkörre szerepel egy függvény, ami a `run_transient` számára definiálja meg az áramkört és egy másik, ami a számunkra kirajzolja.

Ezeket a cellákat a megoldáshoz elég egyszer lefuttatni és nincs szükség módosítani.

````python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

import ahkab
from ahkab import new_ac, run
from ahkab import circuit, printing, time_functions

import SchemDraw as schem
import SchemDraw.elements as elem
````

````python
def run_transient(mycircuit, fun, tstart=0, tstop=1e-2, tstep=1e-5):
    gnd = mycircuit.get_ground_node()
    mycircuit.add_vsource("V_in", n1="n_in", n2=gnd, dc_value=5, ac_value=1, function=fun)
    t = ahkab.new_tran(tstart=tstart, tstop=tstop, tstep=tstep, x0=None)
    r = ahkab.run(mycircuit, an_list=[t])
    time = r['tran']['T']
    source = fun(r['tran']['T'])
    response = r['tran']['Vn_out']
    mycircuit.remove_elem("V_in")
    return time, source, response
````

## Egyszerű aluláteresztő RC-kör

````python
def rc_lowpass():
    c = circuit.Circuit(title="RC-tag")
    c.add_resistor("R1", n1="n_in", n2="n_out", value=600)
    c.add_capacitor("C1", n1="n_out", n2=c.gnd, value=210e-9)
    return c

def draw_rc_lowpass():
    d = schem.Drawing()
    V_in = d.add(elem.SOURCE_SIN, d='up')
    R = d.add(elem.RES, d='right', label='600')
    C = d.add(elem.CAP, d='down', botlabel='210n')
    d.add(elem.LINE, to=V_in.start)
    d.add(elem.LINE, d='right', xy=R.end)
    d.add(elem.DOT, label='a')
    d.add(elem.LINE, d='right', xy=C.end)
    d.add(elem.DOT, label='b')
    d.draw()
    
draw_rc_lowpass()
````

## 4-ed rendű Butterworth-féle aluláteresztő szűrő

````python
def butterworth_lowpass():
    c = circuit.Circuit(title="4-ed rendű Butterworth-féle aluláteresztő szűrő")
    c.add_resistor("R1", n1="n_in", n2="n2", value=600)
    c.add_inductor("L1", n1="n2", n2="n3", value=15e-3)
    c.add_capacitor("C1", n1="n3", n2=c.gnd, value=200e-9)
    c.add_inductor("L2", n1="n3", n2="n_out", value=61e-3)
    c.add_capacitor("C2", n1="n_out", n2=c.gnd, value=150e-9)
    c.add_resistor("R2", n1="n_out", n2=c.gnd, value=1.1e3)
    return c

def draw_butterworth_lowpass():
    d = schem.Drawing()
    V_in = d.add(elem.SOURCE_SIN, d='up')
    R1 = d.add(elem.RES, d='right', label='$R_1$\n600')
    L1 = d.add(elem.INDUCTOR2, d='right', label='$L_1$\n15m')
    C1 = d.add(elem.CAP, d='down', label='$C_1$\n200n')
    L2 = d.add(elem.INDUCTOR2, d='right', xy=C1.start, label='$L_2$\n61m')
    C2 = d.add(elem.CAP, d='down', label='$C_2$\n150n')
    d.add(elem.LINE, d='right', xy=C2.start)
    R2 = d.add(elem.RES, d='down', label='$R_2$\n1.1k')
    for i in range(4): d.add(elem.LINE, d='left')
    d.add(elem.LINE, d='right', xy=R2.start)
    d.add(elem.DOT, label='$V_\mathrm{out}$')
    d.add(elem.LINE, d='right', xy=R2.end)
    d.add(elem.DOT)
    d.draw()
    
draw_butterworth_lowpass()
````

## 4-ed rendű Butterworth-féle sáváteresztő szűrő

````python
def butterworth_bandpass():
    c = circuit.Circuit(title="Butterworth-féle 1kHz-es sáváteresztő szűrő")
    c.add_resistor('R1', 'n_in', 'n2', 50.)
    c.add_inductor('L1', 'n2', 'n3', 12e-3)
    c.add_capacitor('C1', 'n3', 'n4', 2.2e-6)
    c.add_inductor('L2', 'n4', c.gnd, 2e-3)
    c.add_capacitor('C2', 'n4', c.gnd, 13e-6)
    c.add_inductor('L3', 'n4', 'n5', 39e-3)
    c.add_capacitor('C3', 'n5', 'n6', 680e-9)
    c.add_inductor('L4', 'n6', c.gnd, 2e-3)
    c.add_capacitor('C4', 'n6', c.gnd, 13e-6)
    c.add_capacitor('C5', 'n7', 'n_out', 2.2e-6)
    c.add_inductor('L5', 'n6', 'n7', 12e-3)
    c.add_resistor('R2', 'n_out', c.gnd, 50.)
    return c

def draw_butterworth_bandpass():
    d = schem.Drawing()
    V_in = d.add(elem.SOURCE_SIN, d='up')
    R1 = d.add(elem.RES, d='right', label='$R_1$\n50')
    L1 = d.add(elem.INDUCTOR2, d='right', label='$L_1$\n12m')
    C1 = d.add(elem.CAP, d='right', label='$C_1$\n2.2$\mu$')
    L2 = d.add(elem.INDUCTOR2, d='down', label='$L_2$\n2m')
    d.add(elem.LINE, d='right', xy=L2.start)
    C2 = d.add(elem.CAP, d='down', label='$C_2$\n13$\mu$')
    L3 = d.add(elem.INDUCTOR2, d='right', xy=C2.start, label='$L_3$\n39m')
    C3 = d.add(elem.CAP, d='right', label='$C_3$\n680n')
    L4 = d.add(elem.INDUCTOR2, d='down', label='$L_4$\n2m')
    d.add(elem.LINE, d='right', xy=L4.start)
    C4 = d.add(elem.CAP, d='down', label='$C_4$\n13$\mu$')
    L5 = d.add(elem.INDUCTOR2, d='right', xy=C4.start, label='$L_5$\n12m')
    C5 = d.add(elem.CAP, d='right', label='$C_5$\n2.2$\mu$')
    R2 = d.add(elem.RES, d='down', label='$R_2$\n50')
    for i in range(9): d.add(elem.LINE, d='left')
    d.add(elem.LINE, d='right', xy=R2.start)
    d.add(elem.DOT, label='$V_\mathrm{out}$')
    d.add(elem.LINE, d='right', xy=R2.end)
    d.add(elem.DOT)
    d.draw()
    
draw_butterworth_bandpass()
````

# Feladatok

## 1. részfeladat - Mérőjelek szimulációja

Analóg áramkörök kísérleti, szimulált és analitikus vizsgálatakor gyakran alkalmazunk ún. mérőjeleket. 

Írjunk függvényeket, melyek megvalósítják

* a tetszőleges frekvenciájú, amplitúdójú és fázisú szinuszos mérőjelet,
* a tetszőleges periódusidejű és amplitúdójú négyszögjelet,
* az egységimpulzust valamint
* az egységugrást.

A feladat későbbi részeinek megkönnyítése érdekében kövessük azt a szokást, hogy úgy írjuk meg ezeket a mérőjel-függvényeket, hogy akár egyszerre több időpontot is ki tudjanak értékelni, ha ezek a $t$ időpontok `array()`ben vannak eltárolva, és a függvény ezeket így kapja meg egyben.

Ábrázoljuk a mérőjeleket grafikonon.

Milyen mintavételezési frekvencia szükséges az $f$ frekvenciájú szinuszhullám mintavételezéséhez?

válasz helye a mintavételezéssel kapcsolatos kérdésre

### A mérőjeleket megvalósítő függvények

````python
# teszt tömb
tt = np.linspace(0,0.01,100000)
````

A későbbi feladatoknál type errort kaptam (pl. a `run_transient` függvénytől), amikor a $t$ paramétert `float`-ként értelmeztem, és akkor is, ha tömbként / listaként. Így célszerű volt olyan függvényeket definiálni, melyek univerzálisak - mindkét típussal működnek egyszerre.

````python
# színuszos jel
def szin(t,A,w,f):
    return A * np.sin(w*t +f)


# négyszög jel
def negy(t,A,T): 
    # tételezzük fel, hogy a t=0 időpillantban egy f=0 félperiódussal indul a jel
    ido = np.linspace(-0.02,0.02,50001)
    f = 0.0+A*np.sign((np.floor(2*(ido-(T/2))/T)%2)-0.9) # a 0.9 arbitráris, lényeg, hogy 0 és 1 közötti szám legyen
    interpolator = interp1d(ido,f)
    return interpolator(t)


# egységimpulzus
def imp(t):
    # tételezzük fel, hogy t=0-ban van az impulzus
    ido = np.linspace(-0.02,0.02,50001)
    impulzus = np.zeros(50001)
    impulzus[25001] = 1
    interpolator = interp1d(ido,impulzus,'linear')
    return interpolator(t)

    
# egységugrás
def ugr(t):
    # tételezzük fel, hogy t=0-ban történik az ugrás
    return np.ones_like(t)
````

### Ábrázolás

````python
# színuszos jel
szinusz=szin(tt,1,1000*(2*np.pi),0)

# négyszög jel
nsz=negy(tt,1,1/1000)

# egységimpulzus
eimp=imp(tt)

# egységugrás
eugr=ugr(tt)
````

````python
fig, axs = plt.subplots(2, 2,figsize=(8,10))


axs[0, 0].plot(tt, szinusz)
axs[0, 0].set_title("Színuszos jel")

axs[0, 1].plot(tt, nsz)
axs[0, 1].set_title("Négyszög jel")

axs[1, 0].plot(tt, eimp)
axs[1, 0].set_title("Egységimpulzus")

axs[1, 1].plot(tt, eugr)
axs[1, 1].set_title("Egységugrás")


for ax in axs.flat:
    ax.set(xlabel="s", ylabel="V")
````

## 2. rész - Mérőjelekre adott válaszok analízise

Szimuláljuk a mérőjelek hatását a fenti függvényekben definiált áramkörökre:

* RC aluláteresztő szűrő,
* Butterworth-féle aluláteresztő szűrő,
* Butterworth-féle 1kHz-es sáváteresztő szűrő.

Segítségül a hívási szintaxist megmutatjuk az alábbi példán, ahol az RC aluláteresztő szűrő válaszfüggvényét szimuláljuk 1 kHz-es szinuszos gerjesztés esetén.

Minhárom példa áramkör alapparaméterei olyanok, hogy az 1-5 kHz tartományban van a lényegi viselkedés.

A helyes működés érdekében okosan kell a paramétereket megválasztani. Az áramkörök érdekes tartományát figyelembe véve, megfelelő időskálán vizsgáljuk a működést, és a periódikus jeleknek frekvenciáját is helyesen válasszuk meg! Az is előfordulhat, hogy a példában rögzített értékeken módosítani kell például a `run_transient()`ben megadott `tstop` értékén.

A szimulációt elvégző `run_transient()` függvény második argumentumaként várja a bemenő jelet leíró függvényt.
Itt két probléma léphet fel:

Egyik probléma, hogy ez a függvény csak egy argumentumú lehet, az idő az argumentuma. Tehát az első feladatban általánosan elkészített függvényeket, amik további paramétereket is várnak, ügyesen kell ennek a függvénynek átadni. 

Másik, hogy az áramkörszimulátor hibát emel, ha a vizsgáló jelben szakadás vagy ugrás van. Így mind az egységurás mind az impulzus esetén ügyeljünk arra, hogy ezeket a jeleket megfelelően közelítsük. Célszerű a mérőjeleket *folytonossá* alakítani például az `interp1d` osztály segítségével.
Meg kell említeni még, hogy a `run_transient()` kicsivel `tstop` után is kiértékelheti a bemenő jel függvényét, tehát erre fel kell készíteni az implementációt, hogy ilyenkor se emelkedjen hiba.

A szinuszos mérőjel esetén a szakadás miatti probléma nem merül fel, csak a paraméterátadási. Ezt úgy is megoldhatjuk, hogy a kért függvényt ún. `lambda` kifejezéssel helyettesítjük, vagy aki jártas a `python` dekorátorok világában, azokkal is megoldhatja. A lambda kifejezéses megoldásra példa az alábbi:

````python
c = rc_lowpass()
t, s, r = run_transient(c, lambda t: np.sin(1000 * 2 * np.pi * t), tstart=0, tstop=10e-3)

plt.plot(t * 1e3, s)
plt.plot(t * 1e3, r)
plt.xlabel("ms")
plt.ylabel("V")
````

Másik lehetséges megoldás az interpoláció, ami a többi, ugrással rendelkező bemenő jelünk esetében is alkalmazható, az ugrással kapcsolatos problémát és a paraméterátadást is megoldja. Ehhez az `interp1d` függvényt használhatjuk, ami a következőképpen működik.

Ha kiolvassuk a bemenő jelünk értékeit egy elég sűrű pontsorozaton, és átadjuk az `interp1d` függvénynek egymás után a független és függő változókból alkotott értéksorozatot, akkor az elvégzi az ahhoz szükséges számításokat, hogy az eredeti pontsorozat (az alábbi ábrán kék pontok) által képviselt függvényt egyenes szakaszokból álló függvénnyel (piros vonalak) közelítse. Válaszul még nem interpolált értékeket ad vissza, hanem, ahogy az alábbi példa mutatja, egy függvényt, amit meg tudunk hívni a megadott tartományon belüli tetszőleges értékkel az interpolált értékek kiolvasása céljából. Sőt, ilyen értékekből álló listával vagy array-vel is meghívhatjuk, így új értéksorozaton kapjuk meg az interpolált függvény pontjait (piros pontok):

````python
x=[0,0.5,1,1.5,2]; y=[0.05,0.1,0.9,0.95,0.9]  # példaképp ezekre az adatokra interpolálunk
fvinterp=interp1d(x,y)           # a kapott fvinterp valósítja meg az interpolált függvényt
newx=0.75; print(fvinterp(newx))  # próbaképp az fvinterp függvényt itt egy értékkel hívtuk meg
````

````python
newx=np.arange(0,2.001,0.05)  # példa értéksorozat, amin CSAK A SZEMLÉLTETÉS kedvéért
newy=fvinterp(newx)           # kiszámoltatjuk az fvinterp függvénnyel az interpolált értékeket
plt.figure(figsize=(6,4))
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y,'bo')
plt.plot(newx,newy,'r.-');
````

Ez a lehetőség azért jó, mert a `run_transient` függvény a 2. argumentumában szereplő függvényt néha egy értékkel, néha egy array-vel hívja meg. Tehát az interpolált függvényt közvetlen beadhatjuk a `run_transient` függvénynek a korábbi példa szerinti lambda forma helyén. (Persze ekkor a beírt függvény neve után ne írjunk zárójeleket, mert a függvény azonosítóját kell átadni, nem egy lefuttatásának eredményét.)
Természetesen a feladat megoldásához az interpolált függvényt egy megfelelő időskálájú értéksorozatból kell létrehozni.

### Mérőjelek alkalmazása RC aluláteresztő szűrőre

## 3. rész - Átviteli függvény meghatározása

Szimuláljuk a két összetettebb rendszer kimenetét különböző frekvenciájú szinuszos bemenőjelek esetén. Határozzuk meg a kimenet amplitúdóját és fáziseltolódását a frekvencia függvényében. Ügyeljünk arra, hogy a kimeneti jel amplitúdóját a tranziens szakasz eltűnése után kell megmérni!

Ábrázoljuk az átviteli függvényeket és az ábrák elemzésével összegezzük a tapasztaltakat.

### Megoldás

Át kell alakítani a `szin` függvényargumentumát, mert a `curve_fit` csak egy bemeneti paramétert szeret. Ezt a legkönnyebben úgy lehet megoldani, hogy bemeneti paramétereket egy tömbbe fűzzük. Emiatt át fog alakulni a fent függvényben a paraméterek sorrendje, valamint célszerűbb frekvencia paramétert definiálni, mint körfrekvencia paramétert.

````python
def szin_fit(bemenet,A,f):
    t = bemenet[0]
    fr = bemenet[1]
    return A*np.sin(t*2*np.pi*fr+f)
````

A `run_transient` harmadik kimenete a "response", ebből fitteléssel nyejrük ki a frekvenciát és a fázist. Mindként Butterworth-féle szűrőnél ugyanúgy kell majd ezt a műveletet elvégezni, így az egyszerűség kedvéért egy $c$ paraméter formájában megadhatjuk, melyik szűrőről legyen szó.

A `fit` függvény működése a következő:
- Az $fr$ egy tömb, amely tartalmazza a kivizsgálandó frekvenciákat, ezen fut végig egy `for` ciklus
- A `for` ciklus első sorában megnézzük a kiválasztott rendszer válaszát az adott frekvenicás színuszos jelre
- Ezután generálunk egy, a $t$-vel azonos hosszúságú tömböt, hogy a `curve_fit`-ben páronként szerepeljenek a bemeneti adatok (minden $t$ mellett ugyanaz lesz a frekvencia)
- Ezután történik a fittelés, ténylegesen csak a $popt$-ra lesz szükségünk, mert ebben vannak az illesztett paraméterek

````python
from scipy.optimize import curve_fit
````

````python
def fit(fr,c):
    A = np.array([])
    fi = np.array([])
    for i in range(len(fr)):
        t, s, r = run_transient(c,lambda t: szin(t,1,fr[i]*2*np.pi,f), tstart=0, tstop=1e-2)
        frt = fr[i]*np.ones_like(t)
        popt, pcov = curve_fit(szin_fit,[t,frt],r)
        A = np.append(A,popt[0])
        fi = np.append(fi,popt[1])
    return A,fi
````

````python
fr = range(500,5500,50)
````

#### Átviteli függvény meghatározása és ábrázolása a Butterworth-féle aluláteresztő szűrőre

````python
A_low, fi_low = fit(fr, butterworth_lowpass())
````

````python
plt.title("Butterworth-lowpass : Amplitúdó")
plt.plot(np.array(fr),A_low)
plt.xlabel("frekvencia [Hz]")
plt.ylabel("amplitúdó [V]")
plt.grid()
plt.show()

plt.title("Butterworth-lowpass : Fázistolás")
plt.plot(np.array(fr),fi_low)
plt.xlabel("frekvencia [Hz]")
plt.ylabel("fázistolás")
plt.grid()
plt.show()
````

Észrevehetjük, hogy a `curve_fit` néhol negatív értéket adott vissza az amplitúdóra, ézért a $\phi$ értékében $\pi$ nagyságú ugrások jelentkeznek. Az illesztett adatok némi tisztításra szorulnak, erre az alábbi függvényt használjuk.

Gyakorlatilag csak akkor történik ugrás a $\phi$-ben, amikor az amplitúdó előjelet vált.

````python
def tisztit(A,fi):
    a = list(A)
    b = list(fi)
    for i in range(1,len(fi)):
        if (a[i-1]<0) and (a[i]>0):
            a[i-1] = np.abs(a[i-1])
            b[i-1] = b[i-1]-np.pi
        if (a[i-1]>0) and (a[i]<0):
            a[i] = np.abs(a[i])
            b[i] = b[i]+np.pi
        if ((b[i-1]-b[i])>np.pi/2):
            b[i] = b[i]+2*np.pi
        if (b[i]<0):
            b[i] = b[i] + 2*np.pi
        if (b[i-1]<0):
            b[i-1] = b[i-1] + 2*np.pi
    
    for i in range(1,len(fi)):
        if (np.abs(b[i-1]-b[i])>1.97*np.pi):
            for j in range(i):
                b[j] = b[j] - (b[i-1]-b[i])
    return a, b
````

````python
A_l, fi_l = tisztit(A_low,fi_low)
````

````python
plt.title("Butterworth-lowpass : Amplitúdó")
plt.plot(np.array(fr),A_l)
plt.xlabel("frekvencia [Hz]")
plt.ylabel("amplitúdó [V]")
plt.grid()
plt.show()

plt.title("Butterworth-lowpass : Fázistolás")
plt.plot(np.array(fr),fi_l)
plt.xlabel("frekvencia [Hz]")
plt.ylabel("fázistolás")
plt.grid()
plt.show()
````

#### Átviteli függvény meghatározása és ábrázolása a Butterworth-féle sáváteresztő szűrőre

````python
A_band, fi_band = fit(fr, butterworth_bandpass())
````

````python
A_b, fi_b = tisztit(A_band,fi_band)
````

````python
plt.title("Butterworth-bandpass : Amplitúdó")
plt.plot(np.array(fr),A_b)
plt.xlabel("Frekvencia [Hz]")
plt.ylabel("Amplitúdó [V]")
plt.grid()
plt.show()

plt.title("Butterworth-bandpass : Fázistolás") # Erre is működött a tisztító függvény
plt.plot(np.array(fr),fi_b)
plt.xlabel("Frekvencia [Hz]")
plt.ylabel("Fázistolás")
plt.grid()
plt.show()
````

## 4. rész - Az impulzusválasz tulajdonságainak vizsgálata

Szimuláljuk a fent használt sáváteresztő szűrő impulzusra adott válaszát! Határozzuk meg az impulzusválasz Fourier-transzformáltját is. Hogyan viszonyul az impulzusválasz Fourier-transzformáltja az előző feladatban meghatározott átviteli függvényhez?

Mivel a szimuláció időben nem egyeneletes lépésközű adatsort ad, ahhoz, hogy a Fourier-transzformáltját tudjuk venni, át kell mintavételezni egyenletes lépésközzel. Ezt a 2. részfeladatban említett `interp1d` függvénnyel tudjuk megtenni. Most a kiindulási adatsor (kék pontok) nem egyenletes közű, és az egyenközű új idősorozat (piros pontok) nem okvetlen sűrűbb mindenhol. Harmadik eltérés, hogy a valódi jelalakot pontosabban közelítjük, ha az `interp1d` függvénytől az alábbiak szerint harmadrendű spline közelítést kérünk, mert akkor az eredeti adatpontokban simán folytatódó függvényalakot valósít meg. Az spline illesztéssel egy későbbi elméleti anyagrész részletesen fog foglalkozni. Itt elegendő annyit tudni róla, hogy sima függvény esetén pontosabb közelítést ad, mint a lineáris interpoláció. Ugyanakkor ugrással, vagy töréssel rendelkező függvény esetében ezt nem helyes alkalmazni, mert tévesen folytonos deriválttal rendelkező függvényalakot ad, és a töréspont két oldalán oszcillációt okoz. Sima függvény esetére szemléltetésül bemutatunk egy mintaadatsoron elvégzett interpolációt:

````python
x=np.array([0,0.2,0.4,0.7,1,1.5,2,2.6,3.3]); y=np.sin(x)+0.5*np.sin(3*x) # példaképp ezekre az adatokra interpolálunk
plt.figure(figsize=(6,4))
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y,'bo')
fvinterp=interp1d(x,y,"cubic")  # a kapott fvinterp valósítja meg az interpolált függvényt
newx=np.arange(0,3.3,0.3)       # új, egyenközű x értéksorozat
newy=fvinterp(newx)             # kiszámoltatjuk az fvinterp függvénnyel az interpolált értékeket a továbbszámoláshoz
plt.plot(newx,newy,'r.')
finex=np.arange(0,3.3,0.01)       # egy még finomabb felbontású sorozat
finey=fvinterp(finex)             # és a függvényértékek
plt.plot(finex,finey,'r.',ms=1);  # az fvinterp által megvalósított függvényalak kirajzolására
````

Szimuláljuk a fent használt sáváteresztő szűrő impulzusra adott válaszát.



### Sáváteresztő szűrő impulzusválasza

````python
def sav_kobos(t):
    t_sav,s_sav,r_sav = run_transient(butterworth_bandpass(),imp, tstart=0, tstop=10e-3)
    interpolator = interp1d(t_sav,r_sav,'cubic',fill_value = "extrapolate")
    return interpolator(t)
````

````python
eimp = imp(tt)
sk = sav_kobos(tt)
plt.title("Butterworth-bandpass : Válasz impulzusra spline közelítéssel")
plt.plot(tt,sk)
plt.plot(tt,eimp)
plt.ylim([-0.0007,0.0007])
plt.xlabel("idő [s]")
plt.ylabel("amplitúdó [V]")

plt.grid()
plt.show()
````

Mintavételezzük a válaszfüggvényt a fentiek alapján egyenletes időlépéssel, majd határozzuk meg a Fourier-transzformáltját. 

````python
# Sáváteresztő szűrő impulzusválaszának Fourier-transzformáltja
# összhasonlítva a 3. részben kapott átviteli függvénnyel
````

````python
imp_fur = np.fft.fft(sk)
f_fur = np.fft.fftfreq(len(sk),tt[1]-tt[0])
````

````python
plt.title("Impulzusválasz Fourier-transzformáltja")
plt.plot(f_fur,imp_fur)
plt.xlim(0,5000)
plt.xlabel("Frekvencia")
plt.show()
````

````python
plt.title("Impulzusválasz Fourier-transzformáltja - abszolút érték")
plt.plot(f_fur,np.abs(imp_fur))
plt.xlim(0,5000)
plt.xlabel("Frekvencia")
plt.show()
````
