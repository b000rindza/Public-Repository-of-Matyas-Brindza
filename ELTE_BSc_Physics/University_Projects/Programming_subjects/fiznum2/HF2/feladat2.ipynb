# Időbeli folyamatok korrelációjának vizsgálata
Elmélet emlékeztető

Időbeli folyamatok vizsgálatakor érdekes kérdés, hogy két változó "mennyire egyszerre" változik, illetve a változások milyen időbeli eltolással követik egymást. Jelöljön $f(t)$ és $g(t)$ két folytonos, de időben diszkréten és véges módon mintavételezett folyamatot. A két folyamat keresztkorrelációs függvényét, mely a $\tau$ _időeltolás_ függvényében adja meg, hogy a két folyamat mennyire egyszerre változik, a következő módon definiálhatjuk:
$$
R_{fg}(\tau) = \sum_t f(t) \, g(t + \tau) = \sum_t f(t - \tau) g(t) \;.
$$
(Bővebben ld. az elméleti anyagban és az érdeklődőbbeknek: https://en.wikipedia.org/wiki/Cross-correlation)

Gyakran érdemes azt vizsgálni, hogy az átlagukhoz képest mennyire térnek ki hasonlóan, az átlagokat levonva a kovariancia-függvényt kapjuk.
Ha pedig ezt a két jel szórásával lenormáljuk (ugyanis így egyező jelek esetén 1-et kapnánk), akkor kapjuk a normált korrelációs függvényt:
$$
K_{fg}(\tau) = \frac{1}{\sigma_f \sigma_g} \sum_t \left( f(t) - \mu_f \right) \left( g(t + \tau) - \mu_g \right)\;,
$$
ahol $\mu_f, \mu_g$ a két jel átlaga, $\sigma_f, \sigma_g$ pedig a szórásuk.

Belátható, hogy ha a folyamatokat csak egy időben véges intervallumon ismerjük, akkor a korrelációs függvényt csak egy véges hosszúságú $\tau < \tau_{\mathrm{max}}$ időablakkal érdemes meghatározni. A keresztkorrelációs függvény maximumának megkeresésével megtalálhatjuk azt az időeltolást, amely mellett a két időbeli folyamat a leginkább korrellál.

Amennyiben egy időbeli folyamat korrelációját nem egy másik folyamathoz, hanem saját magához képest vizsgáljuk, úgy keresztkorreláció helyett [autokorrelációról](https://en.wikipedia.org/wiki/Autocorrelation) beszélünk. Belátható, hogy az autokorrelációs függvény páros, illetve hogy periodikus jelek autokorrelációja periodikus. Nem periodikus jelek esetében az autokorrelációs függvény $\tau = 0$-tól távolodva lecsengő. A lecsengés üteme függ a vizsgált folyamatoktól. Véletlenszerű zajok és erősen kaotikusan viselkedő rendszerek esetében a lecsengés exponenciálisan gyors (pontosabb az erősen keverő elnevezés), de előfordulnak olyan kaotikus folyamatok is, ahol a korrelációk csak hatványfüggvény szerint csengenek le.

A kereszkorrelációs függvényhez nagyon hasonló módon definiáljuk diszkréten mintavételezett jelek [konvolúcióját](https://en.wikipedia.org/wiki/Convolution):
$$
f(t) \star g(t) = \sum_\tau f(t) \, g(t - \tau)
$$

Lineáris rendszerek esetében korábban már láttuk, hogy a konvolúció művelete segítségével igen könnyen meghatározható a rendszer válasza, ha a [Green-függvény](https://en.wikipedia.org/wiki/Green%27s_function) (impulzusválasz) ismert. A konvolúció művelet Fourier-térben szorzásnak felel meg:

$$
(f \star g)(\omega) = f(\omega) g(\omega)
$$

Az autokorrelációs függvény szintén érdekes kapcsolatban áll a Fourier-transzformációval. A [Wiener-Hincsin-tétel](https://en.wikipedia.org/wiki/Wiener–Khinchin_theorem) kimondja, hogy egy függvény autokorrelációs függvényének Fourier-transzformáltja megegyezik a függvény teljesítményspektrumával, azaz Fourier-transzformáltjának abszolút érték négyzetével:

$$
R_{ff}(\omega) = \left| f(\omega) \right|^2
$$

### 1.a) feladatrész megoldása - jelalakok

A 'boxcar()', 'triang()' és 'square()' függvényeket használjuk a jelek generálásához, melyek bemeneti paraméterként egy integer-t fogadnak el, ami a kívánt jel hosszúsága. Az általam megírt függvények - az 'A' amplitúdó és a 'T' szélesség mellett - egy 't' tömböt vár bemeneti paraméterként, melynek bármely két szomszédos eleme között ugyakkora különbség van.

Ezért a 't' tömb léptékétől fog függni, hogy egy 'T' ideig tartó impulzus mennyi 't'-beli pontnak felel meg, azaz mekkora hosszt adjunk meg a jelgeneráló függvényeknek. Az alábbiakban a 'lenght' változóba kerül az előbbiek alapján a jelhossz, illetve a 'window' változóba kerül a jelgeneráló függvények kimenete.

Ha azonban 'T' kisebb, mint a 't'-ben tárolt időtartam, a megoldás tömbjét fel kell tölteni $0$-kal - ha ugyanolyan hosszú a kimeneti tömb és az eredeti 't' tömb, kényelmesen használhatóak együtt.

Megjegyzés : Feltesszük, hogy $t=0$-ban kezdődnek a jelek. A scipy.signal függvények nem $0$-tól vételez pontokat, ezért egy $0$-val kell kezdeni a végső kimenetet.

```python
# Négyszög alakú impulzus - A amplitúdóval, T szélességgel
def negyimp(t, A, T):
    
    # Mennyi t-beli pont esik a T időtartamba
    length = int(T / np.abs(t[1]-t[0]))
    
    # A jel generálása
    window = A * boxcar(length)
    
    # A jelen kívüli maradék hely feltöltése 0-val
    zero = np.zeros(len(t) - length - 1)
    
    # A résztömbök összefűzése
    return np.hstack( ([0], window, zero ) ).ravel()



# Egyenlő szárú háromszög alakú impulzus - A amplitúdóval, T szélességgel
def isosceles(t, A, T):
    
    # Mennyi t-beli pont esik a T időtartamba
    length = int(T / np.abs(t[1]-t[0]))
    
    # A jel generálása
    window = A * triang(length)
    
    # A jelen kívüli maradék hely feltöltése 0-val
    zero = np.zeros(len(t) - length - 1)
    
    # A résztömbök összefűzése
    return np.hstack( ([0], window, zero ) ).ravel()



# Függőlegesen felfutó, ferdén lefutó háromszögimpulzus - A amplitúdóval, T szélességgel
def haromimp(t,A,T):
    
    # Mennyi t-beli pont esik a T időtartamba
    length = int(T / np.abs(t[1]-t[0]))
    
    # Generálunk egy egyenlő szárú háromszöget, és félbevágjuk
    window0 = A * triang(2*length)
    window = window0[length:]
    
    # A jelen kívüli maradék hely feltöltése 0-val
    zero = np.zeros(len(t) - length - 1)
    
    # A résztömbök összefűzése
    return np.hstack( ([0], window, zero ) ).ravel()



# Négyszögjel - A amplitúdóval, T periódusidővel
def negyszog(t,A,T):

    # A square() függvény alapértelmezett w=1/s körfrekvenciával generál jelet
    # Itt most w = 2*pi/T a körfrekvencia
    # Úgy használjuk, mintha a sin() függvény lenne, csak épp négyszögletes jelet ad
    
    return A * square(2*np.pi * t/T)
```

```python
time_test = np.arange(0,10,0.0005) # teszt tömb az időnek

A, T, T_nsz = 1, 1, 2 # amplitúdó, szélesség és periódus

# Adatok kigenerálása
s_negyimp   = negyimp  (time_test, A, T)
s_isosceles = isosceles(time_test, A, T)
s_fureszimp  = haromimp (time_test, A, T)
s_negyszog  = negyszog (time_test, A, T_nsz)
```

```python
# Az adatok wrap-pelése könnyebb használat érdekében

data_arr  = np.array([s_negyimp, s_isosceles, s_fureszimp, s_negyszog])
label_arr_long = np.array(["Négyszög impulzus", "Egyenlő szárú háromszög impulzus", "Fűrészfog alakú impulzus", "Négyszög alakú periodikus jel"])
label_arr = np.array(["Négyszög imp.", "Egyenlő szárú imp.", "Fűrészfog imp.", "Négyszögjel"])
```

Egy érdekes kísérlet, amiben véletlen folyamatnak, és a véletlen jelek statisztikus tulajdonságainak szerepe van, a Hold-radar kísérlet, melynek során a Holdról visszavert radarjeleket tudtak kimutatni.
(Ez néhány hét időkülönbséggel sikerült amerikai és magyar kutatóknak:
https://hu.wikipedia.org/wiki/Magyar_Hold-radar-kísérlet, https://en.wikipedia.org/wiki/Project_Diana)
A nehézség az volt, hogy a visszavert jel gyengesége miatt egyszerű módon nem tudták a vevők által fogott zajos jelből a visszevert jelet kiszűrni. A jelküldés sokszori ismétlésével és a felfogott jelek összegzésével azonban a jel felerősödött a zajhoz képest.

A legegyszerűbb véletlen folyamat az egymás utáni időpontokban vett egymástól független véletlen értékek sorozata. Ezt is fehér zajnak hívjuk, mert a teljesítmény spektruma egyenletes a maximális frekvenciáig.
Bár a teljesítményspektrum erősen oszcillál, sok zaj-realizációra átlagolva konstanst ad.
(Szintén konstanshoz közelít, ha egy zajsorozatra vesszük, de simítást alkalmazunk.)

Bonyolultabb véletlen folyamatok egyik fajtája a véletlen bolyongás, melyben egy "részecske" valamilyen valószínűségi eloszlás szerint meghatározott lépések megtételével mozog egy, kettő vagy több dimenzióban.
A diszkrét időlépésekben történő bolyongást gyakran a [Brown-mozgás](https://en.wikipedia.org/wiki/Brownian_motion) egyszerű leírására használjuk.
Az egyszerű bolyongás esetén az időfejlődést leíró egyenlet
$$ x(t_{k+1}) = x(t_k) + \xi_k\;,\;\; \mbox{azaz}\;\;\;\; x(t_k) = \sum_{j=0}^{k-1} \xi_j\;, $$
ahol $\xi_k$ a véletlen változó.

Az időben folytonos véletlen folyamatokat ún. sztochasztikus differenciálegyenletekkel írjuk le.
Például a véletlen bolyongás határesete a Wiener-folyamat, melynek egyenlete
$$
\dot x = \xi(t)\;,
$$
ahol $\xi(t)$ a véletlenszerű fehér zajt reprezentálja.

A véletlen folyamatok közül fontosak a stacionáriusak, amelyeknél a felvett $x$ értékek hosszú időre meghatározott eloszlásfüggvénye invariáns az időeltolásra. Ez a tulajdonság az egyszerű bolyongásra és a Wiener-folyamatra nem teljesül, mivel a koordináta értékek eloszlása egyre szélesedik.
A Brown mozgást végző részecske sebességének valószínűségeloszlása azonban stacionárius, ha a sebesség minden határon túli növekedésének a közeg viszkozitása gátat szab.
Ezt folytonos időre írja le az ún. [Ornstein-Uhlenbeck-folyamat](https://en.wikipedia.org/wiki/Ornstein–Uhlenbeck_process), aminek differenciálegyenlete
$$ \dot v = - \gamma v + D\cdot\xi(t)\;,$$
ahol $\gamma$ és $D$ pozitív konstansok, melyek a csillapítás erősségét, illetve a folyamat diffúzivitását adják meg, $\xi(t)$ továbbra is fehér zaj.
Talán könyebben megérthető ennek időben diszkretizált változata:
$$ v(t_{k+1}) = (1-\gamma \Delta t)\cdot v(t_k) + D\cdot\xi_k\;,$$
amiben az első tag szemléletesen a sebesség csillapodását írja le, a második pedig az időlépés alatt a közeg részecskéivel történt ütközések eredő hatását (valójában az utóbbi függ a részecske sebességétől, de ez a hatás átlagolva az első tagba van beleértve).
Érdemes megjegyezni, hogy a sebesség időfejlődése hasonlít a véletlen bolyongáshoz, és azzá is válik a $\gamma=0$ határesetben.

```python
def randomwalk(N):
    # az np.cumsum visszaadja a kumulatív összeget, és az összes közbelső lépést
    k = np.random.normal(size=N)
    return np.cumsum(k)
```

```python
# d)
# A Brown mozgáshoz segítség:
def brownian(N, alpha=0.01, D=1):
    """Brown mozgás"""
    step = np.random.randn(N)
    walk = np.zeros(N)
    for i in range(N-1):
        walk[i+1] = (1- alpha)*walk[i] + D*step[i]
    return walk
```

## 3. részfeladat - Kaotikus rendszer vizsgálata

A logisztikus leképezés az egyik legegyszerűbb kaotikus viselkedést mutató rendszer, melyet a következő iteratív formula definiál (az érdeklődőbbek olvashatnak [róla itt](https://en.wikipedia.org/wiki/Logistic_map)):

$$
x_{t + 1} = r \, x_t \left( 1 - x_t \right)
$$

Az $r$ érték egy ún. kontrollparaméter, amelytől függ, hogy a rendszer milyen viselkedést mutat, pl. hosszú távon periodikussá válik, vagy kaotikus viselkedést mutat. A kaotikus viselkedés $r_c \approx 3.56995$ és a maximális $r_{max}=4$ érték között figyelhető meg. Ezt a kaotikus viselkedést megszakítják olyan intervallumok, u.n. periódikus ablakok, ahol periódikus viselkedés figyelhető meg, amint a kinagyított [bifurkációs diagramon](https://en.wikipedia.org/wiki/Logistic_map#/media/File:Subsection_Bifurcation_Diagram_Logistic_Map.png)
jól látszik. A bifurkációs diagram a paraméter függvényében mutatja, hogy az egyes paraméterértékeknél a rendszer hosszútávú viselkedése során milyen $x$ értékeket jár be.
E rendszer az $r_{max}$ értéknél a legerősebben kaotikus.

A kaotikus rendszerek egy részének érdekes jellemzője, hogy bizonyos kontrollparaméter mellett ún. [intermittens](https://en.wikipedia.org/wiki/Intermittency) viselkedést mutatnak. Ilyenkor a rendszer váratlanul hosszabb vagy rövidebb ideig közel periodikus viselkedést mutat, majd idővel újra beáll a káosz, és ez a kettő váltakozik.
Ilyen a logisztikus leképezésben a periódikus ablakoktól kevéssel balra figyelhető meg, pl. $r=3.828$ értéknél, de $r$ értékét tovább csökkentve $r=3.8$-nál már ez a jelenség lényegében eltűnt.

Feladatok:

a) Írjunk függvényt, mely megvalósítja a fent definiált logisztikus leképezést.
Ábrázoljuk a folyamatot az $r$ kontrollparaméter néhány különböző értéke mellett.
Válasszuk ehhez a fent említett erősen kaotikus és intermittens esetet, és keressünk egy-egy hozzájuk hasonlót!

b) Bővítsük az előbbi vizsgálatokat a korrigált korrelációs függvény meghatározásával, és abrázoljuk, hogyan cseng le a korreláció ezekben az esetekben!

```python
# a)
# A logisztikus leképezést megvalósító függvény
```

```python
def logistic(x0,r,tau):
    x = [x0] # ebből lesz a lista, ami tartalmazza sorra az x-eket
    for i in range(tau-1): # időbeli iteráció
        x.append(r*x[-1]*(1-x[-1])) 
    return np.array(x)
```

```python
# paraméterek
r_t = [3.4, 3.56995, 3.828, 4] # ezt a 4 r-t vizsgáljuk
tau = 150 # ennyi lépés
x0 = 0.005 # kezdőérték ( x[0] )
```

```python
data = [[] for _ in range(int(4))] # előbbiekhez hasonlóan járunk el, itt csak az eljárás rövidítésére szolgál

for i in range(2):
    for j in range(2):
        data0 = logistic(x0,r_t[2*i+j],tau)
        data[2*i+j] = np.copy(data0)
```

```python
fig, axs = plt.subplots(2, 2,figsize=(24,12))
plt.suptitle("LOGISZTIKUS LEKÉPEZÉS",fontsize=30)
fig.tight_layout(pad=6.0)



for i in range(2):
    for j in range(2):
        axs[i,j].plot(np.arange(0,tau,1), data[2*i+j])
        axs[i,j].set_title("r = " + str((r_t[2*i+j])),fontsize=25)



for ax in axs.flat:
    ax.set_xlabel("t", fontsize=22)
    ax.set_ylabel("x", fontsize = 22)
```

```python
# a,b)
# A logisztikus leképezés tulajdonságainak vizsgálata
```

```python
data_corr = [[] for _ in range(int(4))] # hasonlóan végig korreláljuk a 4 esetet

for i in range(2):
    for j in range(2):
        data_corr0 = np.correlate(data[2*i+j],data[2*i+j],"full")
        data_corr[2*i+j] = np.copy(data_corr0)
```

```python
fig, axs = plt.subplots(2, 2,figsize=(24,12))
plt.suptitle("LOGISZTIKUS LEKÉPEZÉS (Korreláció)",fontsize=30)
fig.tight_layout(pad=6.0)



for i in range(2):
    for j in range(2):
        axs[i,j].plot(np.arange(-tau+1,tau,1), data_corr[2*i+j])
        axs[i,j].set_title("r = " + str((r_t[2*i+j])),fontsize=25)



for ax in axs.flat:
    ax.set_xlabel("t", fontsize=22)
    ax.set_ylabel("x", fontsize = 22)
```

Lecsengő korrelációt azért tapasztalunk elsősorban, mert két véges hosszúságú diszkrét adatsort fésülünk át egymáson. Amikor nem érnek egybe (az elején és a végén), akkor az általuk bezárt terület $0$ lesz. A korreláció maximuma várhatóan a $0$ eltolásnál ($\tau$-hoz képest) van.

Az autokorreláció lineárisan lecsengő tendenciát mutat, többnyire csökkenő szélességű "sávban". Néhol "ki-kiszélesedik" a sáv - erre az alábbiakban rátérünk.

Megjegyzés : Ha egy-egy periódus épp egymásba ér, ahogy "csúsztatjuk egymáson" a két adatsort, akkor nagyobb területet zárnak be egymással, így a korrelációban lényeges (látható) növekedést tapasztalunk.

Az felső és az alsó ábrák között észrevehető egy fontos különbség:
- A felső ábráknál a sáv alsó és felső széle is lineárisan cseng le. Ezt csupán azt jelenti, hogy mennek át egymáson a periódusok - csak a $0$-tól jobbra vagy balra tartva egyre kevesebb periódus fed át.
- Az alsó ábráknál gyorsan elvékonyodik a sáv, random módon ugrál le-fel a korreláció a sávban. A bal alsó ábrán megfigyelhető, hogy néhol kiszélesedik a sáv, és hasonló mintát mutat, mint a lebegés effektus. Ez megengedett viselkedés, periódus felbukkanásával megtörténhet.
