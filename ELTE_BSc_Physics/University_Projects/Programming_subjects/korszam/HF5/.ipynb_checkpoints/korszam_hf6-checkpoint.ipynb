# Plotting data

```python
import numpy as np
import scipy
import os
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def slicer(s, c):
    ind = [0]
    string = []
    for i in range(len(s)):
        if s[i] == c:
            ind.append(i)
    ind.append(len(s))
    for i in range(len(ind)-1):
        if i != 0:
            string.append(s[ind[i]+1:ind[i+1]:1])
        if i == 0:
            string.append(s[ind[i]:ind[i+1]:1])
    return string

def filein(name, c):
    a = []
    with open(name) as file:
        lines = file.readlines()
    
    for i in lines:
        a.append([])
        string = slicer(i, c)
        for j in string:
            a[-1].append(int(j))
    return a
```

```python
def cubic(x, a, b):
    return a*(x**3) + b
```

```python
a = filein("data.txt", "\t")
n = []
data = []
for i in a:
    n.append(i[0])
    data.append(i[1::])
```

```python
# average and minimum

dat_avg = [np.array(el).mean() for el in data]
dat_min = [np.array(el).min() for el in data]
```

```python
plt.figure(figsize=(16,12))
plt.title("Dimension of Matrices - Duration of Multiplication")
plt.xlabel("N [1]")
plt.ylabel("Duration [s]")

for i in range(len(data[0])):
    clr = ((i+1)/len(data[0]), i/len(data[0]),0.2)
    plt.plot(n, np.array([el[i] for el in data])/(10**9), ".", label = "Attemp " + str(i+1), color = clr)

plt.legend()
plt.savefig("raw.png")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title("Dimension of Matrices - Duration of Multiplication")
plt.xlabel("N [1]")
plt.ylabel("Duration [s]")

plt.plot(n, np.array([el[0] for el in data])/(10**9), ".", label = "Attempts", color = "#00CCCC")
for i in range(1,len(data[0])-1):
    plt.plot(n, np.array([el[i] for el in data])/(10**9), ".", color = "#00CCCC")

plt.plot(n, np.array(dat_avg)/(10**9), label = "Average", color = "r", linewidth = 2)
plt.plot(n, np.array(dat_min)/(10**9), label = "Minimum", color = "g", linewidth = 2)


plt.legend()
plt.savefig("avg_min.png")
plt.show()
```

```python
popt, pcov = curve_fit(cubic, n, dat_avg)
a_avg = popt[0]
b_avg = popt[1]
popt, pcov = curve_fit(cubic, n, dat_min)
a_min = popt[0]
b_min = popt[1]

sample = np.linspace(n[0],n[-1],1000)
fit_avg = cubic(sample, a_avg, b_avg)
fit_min = cubic(sample, a_min, b_min)
```

```python
plt.figure(figsize=(12,8))
plt.title("Dimension of Matrices - Duration of Multiplication")
plt.xlabel("N [1]")
plt.ylabel("Duration [s]")

plt.plot(n, np.array([el[0] for el in data])/(10**9), ".", label = "Attempts", color = "#00CCCC")
for i in range(1,len(data[0])-1):
    plt.plot(n, np.array([el[i] for el in data])/(10**9), ".", color = "#00CCCC")

plt.plot(sample, fit_avg/(10**9), label = "Average(fit)", color = "r", linewidth = 2)
plt.plot(sample, fit_min/(10**9), label = "Minimum(fit)", color = "g", linewidth = 2)


plt.legend()
plt.savefig("fitted.png")
plt.show()
```
